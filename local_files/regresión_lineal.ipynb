# Regresión Lineal

## Importar Librerias

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar el dataset o conjunto de datos

```
dataset = pd.read_csv('salary.csv')
#print(dataset)
dataset.head()
```

##Crear las variables independiente y dependiente

Utilizando la función loc de la libreria pandas

```
print(dataset.loc[:, ['YearsExperience']])
```

```
print(dataset.loc[:, ['Salary']])
```

Utilizando la función iloc de la libreria pandas

```
print(dataset.iloc[:, :-1])
```

```
print(dataset.iloc[:, 1])
```

Creamos la variable independiente y dependiente

```
x = dataset.loc[:, ['YearsExperience']].values
y = dataset.iloc[:, -1].values

y
```

## División del dataset en conjunto de entrenamiento y conjunto de prueba


Para poder desarrollar machine learning utilizaremos la libreria Sklearn. Si desean más información ir al siguiente link: 
[Página oficial](https://scikit-learn.org/stable/)


Importamos la función train_test_split

```
from sklearn.model_selection import train_test_split
```

Creamos el conjunto de prueba y de entrenamiento

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
y_train
```

## Entrenamiento del modelo de regresión lineal con el conjunto de entrenamiento


Importamos LinearRegression


```
from sklearn.linear_model import LinearRegression
```

Creamos el objeto de la clase LinearRegression

```
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Predicción de los resultados del conjunto de prueba

```
y_pred = regressor.predict(x_test)
```

```
print(y_pred)
```

```
print(y_test)
```

Predicción de valores

```
arr = np.array([[0], [5], [15]])
print(regressor.predict(arr))
```

## Visualización de los resultados del conjunto de entrenamiento

```
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Salario vs Experiencia (Conjunto de entrenamiento)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```

## Visualización de los resultados del conjunto de prueba


```
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Salario vs Experiencia (Conjunto de prueba)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```
